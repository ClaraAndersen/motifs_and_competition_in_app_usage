# UsedTime and sleeptime of individuals


```python
data = pd.read_csv('sorted_coded_data_with_merged_events_10_min.csv',index_col=False)
```

```python
data = data.drop(columns=['Unnamed: 0'])
data
```

```python
data.columns
```

Let's first take a look at the median timespend of all users,min and max

```python
data.time_spend.median()
```

```python
print("min time_spend: ",data.time_spend.min())
print("max time_spend: ",data.time_spend.max())
```

```python
df = pd.DataFrame({'user_id' : range(0, 10000)})
df['median_spend_time']=data.groupby(['user_id']).agg({'time_spend': ['median']})
```

```python
df
```

```python
fig = px.histogram(df, x="median_spend_time",opacity=0.8,title='distribution of time spend of each app individualy',nbins=65).update_xaxes(categoryorder='total ascending')
fig.show()
```

```python
df['median_sleep_time']=data[data.sleep_time_modified>0].groupby(['user_id']).agg({'sleep_time_modified': ['median']})
```

```python
df['median_sleep_time_min'] = df['median_sleep_time']/60
```

```python
df.describe()
```

```python
np.arange(0,60,5)
```

```python
#fig1 = px.histogram(df, x="median_sleep_time_min",opacity=0.8,log_y=True,title='distribution of sleep time of each user individualy',nbins=100).update_xaxes(categoryorder='total ascending')
counts, bins = np.histogram(df.median_sleep_time_min, bins=range(0, 560, 10))
bins = 0.5 * (bins[:-1] + bins[1:])
fig1 = px.bar(x=bins, y=counts, log_y=True,labels={'x':'median_sleep_time_min', 'y':'counts'})
fig1.show()
```
