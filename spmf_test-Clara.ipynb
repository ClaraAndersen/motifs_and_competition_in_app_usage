# MaxSP
The algorithm focuses on identifying the maximal patterns. Since there is so many high frequent short patterns this can be a way to find the most interesting patterns.

```python
spmf = Spmf("PrefixSpan", input_filename="/home/s164574/motifs_and_competition_in_app_usage/test_4users.txt",
            output_filename="/home/s164574/motifs_and_competition_in_app_usage/output_maxSP.txt",
            spmf_bin_location_dir="/home/s164574/motifs_and_competition_in_app_usage",
            arguments=[0.25,5])
spmf.run()
print(spmf.to_pandas_dataframe(pickle=True))
spmf.to_csv("output.csv")
```
