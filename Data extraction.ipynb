## Extracting data files from zip 
a folder with all parquets file

```python
from zipfile import ZipFile
import pandas as pd
import os
```

```python
# specifying the zip file name
file_name = "data_for_LongClara.zip"
  
# opening the zip file in READ mode
with ZipFile(file_name, 'r') as zip:
    # printing all the contents of the zip file
    zip.printdir()
  
    # extracting all the files
    print('Extracting all the files now...')
    zip.extractall()
    print('Done!')
```

## Create all datas into dataframe

```python
directory = 'data_for_LongClara/'
#list the files
filelist = os.listdir(directory) 

#read them into pandas
df_list = [pd.read_parquet('data_for_LongClara/'+ file) for file in filelist if file.endswith(".parquet")] 

#concatenate them together
UK = pd.concat(df_list)

#for file in filelist: 
#    if file.endswith(".parquet"):
#        print(file)
#        pd.read_parquet('data_for_LongClara/'+ file)
        
#    else:
#        continue

```
